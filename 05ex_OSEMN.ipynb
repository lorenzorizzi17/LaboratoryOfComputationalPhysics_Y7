4\. load the binary file named *credit_card.dat* and convert the data into the real credit-card number.
Each line correspond to a credit card number.
Each character is composed by 6 bit (even the space) and the last 4 bit are just a padding

**hint**: use the `chr()` function to convert a number to a char

```python
credit_card = []

with open("./credit_card.dat", mode = "r") as file:
    for line in file:
        if(len(line) == 6*19+4+1):
            splitted = [line[6*i:6*i+6] for i in range(0,20)]
            chars = list(map(lambda x: chr(int(x,2)), splitted))
            credit_card.append("".join(chars[:-1]))
print(credit_card)
        
```

6\. Load the file from this url: [https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1](https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1) with Pandas. 
+ Explore the data (see the info of the data)
+ Draw the istogram of the 'class' field. Decribe wath yuou see


```python
import urllib.request
import csv
url ='https://www.dropbox.com/s/7u3lm737ogbqsg8/mushrooms_categorized.csv?dl=1 '
with urllib.request.urlopen(url) as data_file:
    data=pd.read_csv(data_file,skiprows=range(1,1))

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (10,8))
ax.hist(data["class"].values, bins = np.linspace(-0.5, 1.5, 3))
ax.set_title("Occurencies of class type in the dataset")
plt.show()
```

7\. Load the remote file [https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1](https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1) with Pandas and plot a scatter plot all possible combination of the following fields:
    
  + features_1
  + features_2
  + features_3

```python
import urllib.request
import csv
url ='https://www.dropbox.com/s/vkl89yce7xjdq4n/regression_generated.csv?dl=1'
with urllib.request.urlopen(url) as data_file:
    data=pd.read_csv(data_file,skiprows=range(1,1))

import matplotlib.pyplot as plt

fig, ax = plt.subplots(ncols = 2, nrows = 2, figsize = (10,8))
ax[0,0].scatter(data[data.columns[0]].values, data[data.columns[1]].values,  s = 5, alpha = 0.3)
ax[0,0].set_title("feature1 vs feature2")

ax[0,1].scatter(data[data.columns[1]].values, data[data.columns[2]].values, s = 5, alpha = 0.3)
ax[0,1].set_title("feature2 vs feature3")

ax[1,0].scatter(data[data.columns[0]].values, data[data.columns[2]].values, s = 5, alpha = 0.3)
ax[1,0].set_title("feature1 vs feature3")
plt.show()
```
