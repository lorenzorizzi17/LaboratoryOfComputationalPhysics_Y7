1\. **PCA on 3D dataset**

* Generate a dataset with 3 features each with N entries (N being ${\cal O}(1000)$). With $N(\mu,\sigma)$ the normali distribution with mean $\mu$ and $\sigma$  standard deviation, generate the 3 variables $x_{1,2,3}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
* Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result
* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained.
* Redefine the data in the basis yielded by the PCA procedure
* Plot the data points in the original and the new coordiantes as a set of scatter plots. Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg as la
N = 1000
first_feature = np.random.normal(0,1, (N,1))
second_feature = first_feature + np.random.normal(0,3, (N,1))
third_feature = 2*first_feature + second_feature
dataset_np = np.concatenate((np.concatenate((first_feature, second_feature), axis=1),third_feature),axis = 1)
dataset = pd.DataFrame(np.concatenate((np.concatenate((first_feature, second_feature), axis=1),third_feature),axis = 1), columns = ["x1", "x2", "x3"])
#now find the eigenvalues and eigenvectors
cov_matrix = np.cov(dataset_np.T)
l, V = la.eig(cov_matrix)
print("The eigenvalues of the covariance matrix are", np.real_if_close(l))
print("The first eigenvector is: " , V[:,0])
print("The second eigenvector is: " , V[:,1])
print("The third eigenvector is: " , V[:,2])
#now finding the same eigen-quantities with SVD
U, spectrum, Vt = la.svd(cov_matrix) #three numpy arrays in return 
print("Using SVD, the eigenvalues are: ", spectrum, " and they are equal? ", np.allclose(spectrum, np.flip(np.sort(np.real_if_close(l)) ) ))
print("The first eigenvector is: " , U[:,0])
print("The second eigenvector is: " , U[:,1])
print("The third eigenvector is: " , U[:,2])

print("Percentage of higher eigenvalue: ",100* (np.real_if_close(l[0]))/np.real_if_close(np.sum(l)))
print("Percentage of median eigenvalue: ",100* (np.real_if_close(l[2]))/np.real_if_close(np.sum(l)))
print("Percentage of smallest eigenvalue: ",100* (np.real_if_close(l[1]))/np.real_if_close(np.sum(l)))

#So we have to get rid of the combination -0.81649658 * x1 - 0.40824829*x2 + 0.40824829*x3
#We build a new dataset with only the two combinations given by the highest eigenvalues
# oppure ruota
new_dataset = np.concatenate((V[0,0]*first_feature+V[0,1]*second_feature+V[0,2]*third_feature, V[1,0]*first_feature+V[1,1]*second_feature+V[1,2]*third_feature),axis=1)

fig, ax = plt.subplots(nrows = 2, ncols =3 , figsize=(16, 10))
ax[0,0].set_xlim(-10.,10)
ax[0,0].set_ylim(-10,10)
ax[0,0].set_title("x1 against x2 (old coord)")
ax[0,0].scatter(x=dataset_np[:,0],y = dataset_np[:,1], marker='o', c='r', edgecolor='r',lw = 0.1)

ax[0,1].set_xlim(-15.,15)
ax[0,1].set_ylim(-15,15)
ax[0,1].set_title("x1 against x3 (old coord)")
ax[0,1].scatter(x=dataset_np[:,0],y = dataset_np[:,2], marker='o', c='b', edgecolor='r',lw = 0.1)

ax[0,2].set_xlim(-15.,15)
ax[0,2].set_ylim(-15,15)
ax[0,2].set_title("x2 against x3 (old coord)")
ax[0,2].scatter(x=dataset_np[:,1],y = dataset_np[:,2], marker='o', c='g', edgecolor='r',lw = 0.1)


rotated_dataset = np.array([np.dot(V.T,dataset_np[i,:]) for i in range(N)])

ax[1,0].set_xlim(-15.,15)
ax[1,0].set_ylim(-15,15)
ax[1,0].set_title("x1 against x2 (new coord)")
ax[1,0].scatter(x=rotated_dataset[:,0],y = rotated_dataset[:,1], marker='o', c='r', edgecolor='r',lw = 0.1)

ax[1,1].set_xlim(-10.,10)
ax[1,1].set_ylim(-10,10)
ax[1,1].set_title("x1 against x3 (new coord)")
ax[1,1].scatter(x=rotated_dataset[:,0],y = rotated_dataset[:,2], marker='o', c='b', edgecolor='r',lw = 0.1)


ax[1,2].set_xlim(-10.,10)
ax[1,2].set_ylim(-10,10)
ax[1,2].set_title("x2 against x3 (new coord)")
ax[1,2].scatter(x=rotated_dataset[:,1],y = rotated_dataset[:,2], marker='o', c='g', edgecolor='r',lw = 0.1)


```
