```python
import numpy as np
import pandas as pd
import scipy.linalg as la
import matplotlib.pyplot as plt

N = 1000


x1 = np.random.normal(0,1,size= N)
x2 = x1 + np.random.normal(0,3, size = N)
x3 = 2*x1 + x2
df = pd.DataFrame({"x1" : x1, "x2": x2, "x3": x3})


cov_matrix = np.cov(df.T)
print("Covariance matrix: \n", cov_matrix, "\n")

l, V = la.eig(cov_matrix)
for i in range(len(l)):
    print("Eigenvalue (EIG) : ", np.real_if_close(l[i]), ", eigenvector: ", V[:,i])
print("\n")
U, spectrum, Vt = la.svd(cov_matrix)
for i in range(len(spectrum)):
    print("Eigenvalue (EIG) : ", np.real_if_close(spectrum[i]), ", eigenvector: ", U[:,i])

print("\n")

smallest_eigenvalue = l[np.argmin(l)]
ratio = (np.trace(cov_matrix) - smallest_eigenvalue )/ np.trace(cov_matrix)
print(f"Removing the smallest eigenvalue, we'll still be able to explain {np.real_if_close(100*ratio):.1f} %", )
second_smallest = np.sort( np.real_if_close(l) )[1]
ratio2 = (np.trace(cov_matrix) - smallest_eigenvalue - second_smallest )/ np.trace(cov_matrix)
print(f"Removing also the second to last smallest eigenvalue, we'll still be able to explain {np.real_if_close(100*ratio2):.2f} %")
print("\n")
rotated_data = pd.DataFrame((np.dot(V.T, df.T)).T, columns = ["y1", "y2", "y3"])
print("Rotated dataset: \n", rotated_data)


fig, ax = plt.subplots(ncols = 3, nrows = 2, figsize = (18,10))
ax[0,0].scatter(df["x1"], df["x2"], c = "r")
ax[0,0].set_title("Original data")
ax[0,0].set_xlabel("x1")
ax[0,0].set_ylabel("x2")
ax[0,0].set_xlim(-10,10)
ax[0,0].set_ylim(-10,10)

ax[0,1].scatter(df["x1"], df["x3"], c = "g")
ax[0,1].set_title("Original data")
ax[0,1].set_xlabel("x1")
ax[0,1].set_ylabel("x3")
ax[0,1].set_xlim(-10,10)
ax[0,1].set_ylim(-10,10)


ax[0,2].scatter(df["x2"], df["x3"], c = "b")
ax[0,2].set_title("Original data")
ax[0,2].set_xlabel("x2")
ax[0,2].set_ylabel("x3")
ax[0,2].set_xlim(-10,10)
ax[0,2].set_ylim(-10,10)

ax[1,0].scatter(rotated_data["y1"], rotated_data["y2"], c = "r")
ax[1,0].set_title("Rotated data")
ax[1,0].set_xlabel("y1")
ax[1,0].set_ylabel("y2")
ax[1,0].set_xlim(-10,10)
ax[1,0].set_ylim(-10,10)

ax[1,1].scatter(rotated_data["y1"], rotated_data["y3"], c = "g")
ax[1,1].set_title("Rotated data")
ax[1,1].set_xlabel("y1")
ax[1,1].set_ylabel("y3")
ax[1,1].set_xlim(-10,10)
ax[1,1].set_ylim(-10,10)

ax[1,2].scatter(rotated_data["y2"], rotated_data["y3"], c = "b")
ax[1,2].set_title("Rotated data")
ax[1,2].set_xlabel("y2")
ax[1,2].set_ylabel("y3")
ax[1,2].set_xlim(-10,10)
ax[1,2].set_ylim(-10,10)

```

2\. **Curve fitting of temperature in Alaska** 

The temperature extremes in Alaska for each month, starting in January, are given by (in degrees Celcius):

max:  17,  19,  21,  28,  33,  38, 37,  37,  31,  23,  19,  18

min: -62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58

* Plot these temperature extremes.
* Define a function that can describe min and max temperatures. 
* Fit this function to the data with scipy.optimize.curve_fit().
* Plot the result. Is the fit reasonable? If not, why?
* Is the time offset for min and max temperatures the same within the fit accuracy?

```python
import pandas as pd
from scipy.optimize import curve_fit
from math import pi

max_temp = np.array([17,  19,  21,  28,  33,  38, 37,  37,  31,  23,  19,  18])
min_temp = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
month = np.arange(0,12) + 1
fig, ax = plt.subplots(figsize=(10,8))

ax.plot(month, max_temp, "ro")
ax.plot(month, min_temp, "bo")

def f(x : float, A: float, omega: float, phi : float, B : float):
    return A*np.cos(omega*x+phi) + B

par_min, cov_min = curve_fit(f, month, min_temp, p0 = [(np.max(min_temp)-np.min(min_temp))/2, 2*pi/12, pi, np.mean(min_temp)])
par_max, cov_max = curve_fit(f, month, max_temp, p0 = [(np.max(max_temp)-np.min(max_temp))/2, 2*pi/12, pi, np.mean(max_temp)])

support = np.linspace(1,12,100)

ax.plot(support, f(support, par_min[0], par_min[1], par_min[2], par_min[3]))
ax.plot(support, f(support, par_max[0], par_max[1], par_max[2], par_max[3]))

print(par_min[2], np.sqrt(cov_min[2,2]))
print(par_max[2], np.sqrt(cov_max[2,2]))
```

3\. **2D minimization of a six-hump camelback function**

$$
f(x,y) = \left(4-2.1x^2+\frac{x^4}{3} \right) x^2 +xy + (4y^2 -4)y^2
$$

has multiple global and local minima. Find the global minima of this function.

Hints:

* Variables can be restricted to $-2 < x < 2$ and $-1 < y < 1$.
* Use numpy.meshgrid() and pylab.imshow() to find visually the regions.
* Use scipy.optimize.minimize(), optionally trying out several of its methods.

How many global minima are there, and what is the function value at those points? What happens for an initial guess of $(x, y) = (0, 0)$ ?


```python
from scipy.optimize import minimize

def f(l):
    x = l[0]
    y = l[1] 
    return (4 - 2.1*x*x + pow(x,4)/3)*x*x + x*y + (4*y*y -4)*y*y 

X = np.linspace(-2,2,200)
Y = np.linspace(-1,1,200)
Z = (np.meshgrid(X,Y))

fig, ax = plt.subplots(figsize=(12,6))
im = ax.imshow(f(Z),cmap = "inferno", extent = ([-2,2,-1,1]))
fig.colorbar(im)
ax.set_title("Plot of the 2d function")


def f_opt(x, y):
    return (4 - 2.1*x*x + pow(x,4)/3)*x*x + x*y + (4*y*y -4)*y*y 

# to find all minima,
N = 10
initial_guess = np.array([(i/10,j/10) for i in range(-2*N,2*N,1) for j in range(-1*N,1*N,1)] )
minima = [minimize(f, x0 = initial_guess[i]).x for i in range(len(initial_guess))]

for i in range(len(minima)):
    ax.scatter(initial_guess[i][0],initial_guess[i][1], c = "y", s = 1)
    ax.scatter(minima[i][0], minima[i][1], c = "g")
ax.scatter(minima[:-1][0],minima[:-1][1],c = "g", label ="converging minima points")
ax.scatter(initial_guess[0][0],initial_guess[0][1], c = "y", label ="starting points")
ax.legend(loc = "best")

fig.tight_layout()
plt.show()
```
